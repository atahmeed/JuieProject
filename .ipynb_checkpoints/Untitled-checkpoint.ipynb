# Model Training

```python
# Import KMeans
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split

#train test split
x_train,x_test,y_train,y_test = train_test_split(xs,ys,test_size = 0.25,random_state=42)

# Create a KMeans instance with 3 clusters: model
model = KMeans(n_clusters=2)

# Fit model to points
model.fit(x_train)

# Determine the cluster labels of new_points: labels
labels = model.predict(x_test)

```

```python
predicted = np.empty([labels.shape[0]],dtype=str)
actual = np.empty([y_test.shape[0]],dtype=str)

predicted[labels==0] = "Male"
predicted[labels==1] = "Female"
# predicted = predicted.reshape((-1,1))
actual[y_test==1] = "Male"
actual[y_test==2] = "Female"
# actual = actual.reshape((-1,1))

# Create a DataFrame with labels and varieties as columns: df
df = pd.DataFrame({'predicted': predicted, 'actual': actual})

# Create crosstab: ct
ct = pd.crosstab(df['actual'],df['predicted'])

# Display ct
print(ct)
```
